```python
import networkx as nx
import pandas as pd
```

```python
df=pd.read_csv('arcs.txt', sep=' ', names=['n1','n2','Weight'])
G=nx.from_pandas_edgelist(df,'n1','n2',create_using=nx.DiGraph)
k_pos=nx.spring_layout(G, k=2.0)
nx.draw_networkx(G,k_pos)
```

```python
G.edges()
```

```python
G.nodes()
```

```python
G.number_of_edges()
```

```python
G.number_of_nodes()
```
